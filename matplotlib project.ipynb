```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Data.csv")
profitlist = df ['total_profit'].tolist()
monthlist = df ['month_number'].tolist()
plt.plot(monthlist,profitlist,label = 'last year month_wise profit data')
plt.xlabel('Month Number')
plt.ylabel('Profit in Doller')
plt.xticks(monthlist)
plt.title('Company Profit per Month ')
plt.yticks([100000,200000,300000,400000,50000])
plt.show()
```
